## MaxQuantOutput class

Instead of handling the files manually in a MQ folder, e.g. like

```python
all_files = io.search_files(path=w_folder.value, query='')
all_files
```

```python
mq_output = mq.MaxQuantOutput(w_folder.value)
mq_output
```
